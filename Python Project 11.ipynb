# Employee Performance Analyzer¶

```python
# EMPLOYEE CLASS 
class Employee:
    
    def __init__(self, emp_id, name, department, role, attendance,
                 tasks_completed, work_quality, behavior,
                 teamwork, communication, deadlines_met):
        
        self.emp_id = emp_id                  ### all employee attributes.also called instance variable
        self.name = name
        self.department = department
        self.role = role
        self.attendance = attendance
        self.tasks_completed = tasks_completed
        self.work_quality = work_quality
        self.behavior = behavior
        self.teamwork = teamwork
        self.communication = communication
        self.deadlines_met = deadlines_met
        
 ## automatic performance calculation inside constructor
        
        self.performance_score = self.calculate_performance()
        self.rating = self.get_rating()

    # CALCULATE performance(metthod calculated weighted performance score.
    def calculate_performance(self):
        score = (
            (self.attendance * 0.2) +
            (self.tasks_completed * 0.2) +
            (self.work_quality * 0.2) +
            (self.behavior * 0.1) +
            (self.teamwork * 0.1) +
            (self.communication * 0.1) +
            (self.deadlines_met * 0.1)
        )
        return score

    # Performance rating(This method assign performance category,)
    def get_rating(self):
        if self.performance_score >= 85:
            return "Excellent"
        elif self.performance_score >= 70:
            return "Good"
        elif self.performance_score >= 50:
            return "Average"
        else:
            return "Poor"

    #  Promotion eligibility(It determine whether employee should be promoted or not.)
    def promotion_status(self):
        if self.performance_score >= 80:
            return "Eligible for Promotion"
        elif self.performance_score >= 60:
            return "Needs Improvement"
        else:
            return "Not Eligible"

    # DISPLAY DETAILS (This method prints all employee information.)
    def display_details(self):
        print(f"\nEmployee ID: {self.emp_id}")
        print(f"Name: {self.name}")
        print(f"Department: {self.department}")
        print(f"Role: {self.role}")
        print(f"Attendance: {self.attendance}")
        print(f"Tasks Completed: {self.tasks_completed}")
        print(f"Work Quality: {self.work_quality}")
        print(f"Behaviour: {self.behavior}")
        print(f"Teamwork: {self.teamwork}")
        print(f"Communication: {self.communication}")
        print(f"Deadlines Met: {self.deadlines_met}")
        print(f"Performance Score: {round(self.performance_score, 2)}")
        print(f"Rating: {self.rating}")
        print(f"Promotion Status: {self.promotion_status()}")
        print("-------------------------------------------")


#  PERFORMANCE ANALYZER 
# This class manages all employees and stores it in a list.)
class PerformanceAnalyzer:
    
    def __init__(self):
        self.employee_list = []

    #  ADD EMPLOYEE WITH UNIQUE ID CHECK (befor adding new employee it checks duplicate ID.)
    def add_employee(self, emp):
        if any(e.emp_id == emp.emp_id for e in self.employee_list):
            print(f"Error: Employee ID '{emp.emp_id}' already exists! Employee not added.")
            return False
        self.employee_list.append(emp)
        print("Employee Added Successfully!")
        return True

    # show all employees(loop through employee list if present display or no availabilityof emp)
    def show_all_employees(self):
        if not self.employee_list:
            print("No employee data available")
        else:
            for emp in self.employee_list:
                emp.display_details()

    # average performance(calculate average score of all employee.)
    def average_performance(self):
        if not self.employee_list:
            return 0
        return sum(emp.performance_score for emp in self.employee_list) / len(self.employee_list)

    # best employee(this method finds employee with highest score.)
    def best_employee(self):
        if not self.employee_list:
            return None
        return max(self.employee_list, key=lambda emp: emp.performance_score)

    # low performers(employee with lowest score.)
    def low_performers(self):
        return [emp for emp in self.employee_list if emp.performance_score < 50]


# Menu function(This function display program options.)
def menu():
    print("\n========== EMPLOYEE PERFORMANCE SYSTEM ==========")
    print("1. Add Employee")
    print("2. Show All Employees")
    print("3. Show Average Performance")
    print("4. Show Best Employee")
    print("5. Show Low Performers")
    print("6. Exit")


# Main Programm((creating performanceanalyzer object here we create object of analyzer class ,This object manages all employees.)
analyzer = PerformanceAnalyzer()

while True:   #this programm run until user select exit
    menu()
    
    try:    # (select option from menu)
        choice = int(input("Enter your choice: "))
    except:
        print("Enter valid number!")
        continue
    
    #  ADD Employee (select option 1program ask Id name,role,etc)
    if choice == 1:
        #  Unique ID check loop 
        while True:
            emp_id = input("Enter Employee ID: ")
            if any(e.emp_id == emp_id for e in analyzer.employee_list):
                print(f"Error: Employee ID '{emp_id}' already exists! Please enter a unique ID.")
            else:
                break

        name = input("Enter Name: ")
        department = input("Enter Department: ")
        role = input("Enter Role: ")
        
        try:
            attendance = float(input("Attendance (%): "))
            tasks = float(input("Tasks Completed: "))
            work_quality = float(input("Work Quality Score: "))
            behavior = float(input("Behaviour Score: "))
            teamwork = float(input("Teamwork Score: "))
            communication = float(input("Communication Score: "))
            deadlines = float(input("Deadlines Met Score: "))
        except:
            print("Enter numeric values only!")
            continue

        #(then object created:
        emp = Employee(emp_id, name, department, role,  # 
                       attendance, tasks, work_quality, # 
                       behavior, teamwork, communication,
                       deadlines)
        #(then added
        analyzer.add_employee(emp)

    #  SHOW ALL EMPLOYEES 
    elif choice == 2:
        analyzer.show_all_employees()

    #  AVERAGE PERFORMANCE 
    elif choice == 3:
        avg = analyzer.average_performance()
        print("Average Performance Score:", round(avg, 2))

    #  BEST EMPLOYEE 
    elif choice == 4:
        best = analyzer.best_employee()
        if best:
            print("\nBest Employee Details:")
            best.display_details()
        else:
            print("No employee data available")

    #  LOW PERFORMERS 
    elif choice == 5:
        low = analyzer.low_performers()
        if not low:
            print("No low performers")
        else:
            print("\nLow Performers:")
            for emp in low:
                emp.display_details()

    #  Exit of programme. (stop loops and end program
    elif choice == 6:
        print("Program Ended")
        break

    else:
        print("Invalid Choice")
```
